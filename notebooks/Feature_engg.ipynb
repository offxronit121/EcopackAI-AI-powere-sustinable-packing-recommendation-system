```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

materials = pd.read_csv('/content/materials_cleaned.csv')
```

```python
materials.columns
```

```python
# Strength buckets for readability
def strength_bucket(mpa):
    if mpa < 20:
        return "Low"
    elif mpa < 50:
        return "Medium"
    else:
        return "High"

materials['strength_level'] = materials['strength_mpa'].apply(strength_bucket)

# Check result
materials[['strength_mpa', 'strength_level']].head()
```

```python
materials['strength_level'].head()
```

```python
# Normalize Co2, Cost and Strength fields to 0-1 scale
scaler = MinMaxScaler()

materials[["co2_norm","cost_norm","strength_norm"]] = scaler.fit_transform(materials[["co2_emission_per_kg","cost_inr_per_kg","strength_mpa"]])
```
